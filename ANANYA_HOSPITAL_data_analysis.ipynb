```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("bolpurKRnew_hospitalfinal_updated.csv")


if "admission_date" in df.columns:
    df["admission_date"] = pd.to_datetime(df["admission_date"])  
    df["month"] = df["admission_date"].dt.month_name() 


monthly_counts = df["month"].value_counts()


month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
monthly_counts = monthly_counts.reindex(month_order, fill_value=0)



plt.figure(figsize=(10, 5))
plt.bar(monthly_counts.index, monthly_counts.values, color='pink')
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Number of Patients Admitted")
plt.title("Monthly Patient Admissions")
plt.show()
```

# Insights from the Graph

1. Monthly Trends:

     a> Patient admissions remain high and stable from January to October.
     
     b> Highest number of patient is admit in may.
     
     c> There is a sharp decline in November and December.
 
2. Possible Reasons for the Drop (Nov-Dec):
   
    a> Holiday Effect: Patients may avoid hospitals due to festivals and vacations.

3. Resource Management Implications:
     
    a> Hospitals should maintain steady resources (staff, beds, medicines) from January to October.
    
    b> In November and December, they may optimize operations, adjust staffing, or conduct maintenance.
